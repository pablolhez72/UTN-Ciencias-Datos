```python
# Importar librerías necesarias
import seaborn as sns
import pandas as pd
import warnings
from autoviz.AutoViz_Class import AutoViz_Class

```

```python
# Ignorar todos los warnings
warnings.filterwarnings("ignore")

# 1. Cargar dataset
df=pd.read_csv("/content/Student_Habits.csv", sep=',', low_memory=False)

# Visualizar primeras filas
print(df.head())
```

```python


# Configurar AutoViz
from autoviz.AutoViz_Class import AutoViz_Class
AV = AutoViz_Class()
df_resultado = AV.AutoViz(
filename="",
sep=",",
depVar="exam_score",
dfte=df,
header=0,
verbose=1,
chart_format="bokeh" # Opcional: gráficos interactivos
)
depVar="exam_score",
dfte=df,8
header=0,
verbose=1,
chart_format="bokeh" # Opcional: gráficos interactivos
```
