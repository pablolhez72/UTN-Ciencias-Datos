```python
# Python-for-Beginners   https://lms.simplilearn.com/courses/4214/Python-for-Beginners/syllabus
from matplotlib import pylab
print("pilab")
import numpy as np
# se usa numpy para generar datos aleatorios
x=np.linspace(0,12,25)
y=x*x+2
print(x)
print(y)
```

```python
from matplotlib import pyplot as plt
fig = plt.figure()
axis = fig.add_axes([0.5, 0.1, 0.8, 0.8])
axis.plot(x,y,'r')
```

```python
fig,axes = plt.subplots()
axes.set_title("title bla bla")
ax.legend(["label1","label2"])
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.plot(x, x**2)
axes.plot(x, x**3)
axes.legend(["y=x**2","y=x**3"], loc=2)# loc=2 es la esquina 2 del rectangulo del grafico
```

```python
fig,axes = plt.subplots()
axes.plot(x,x+1, color='red')
axes.plot(x,x+2, color='blue')
axes.plot(x,x+3, color='green')
```

```python
fig,axes = plt.subplots(dpi=100)
axes.plot(x,x+1,color='blue', lw=2, linestyle='-')
axes.plot(x,x+2,color='blue', lw=2, linestyle='-.')
axes.plot(x,x+3,color='blue', lw=2, linestyle=':')
```

```python
fig,axes = plt.subplots(dpi=100)
axes.plot(x,x+1,color='blue', marker='o')
axes.plot(x,x+2,color='blue', marker='+')
axes.plot(x,x+3,color='blue', marker='s')
axes.plot(x,x+4,color='blue', marker='1')
axes.plot(x,x+5,color='blue', marker='o',markersize=2)
axes.plot(x,x+6,color='blue', marker='s',markerfacecolor='red')
```

```python
fig,Axes = plt.subplots(1,2,figsize=(10,5))
Axes[0].plot(x, x**2, x , x**3, lw=2)
#Axes[0].grid(True)

Axes[1].plot(x, x**2, x , x**3, lw=2)
Axes[1].set_ylim([0,60])
Axes[1].set_xlim([2,5])
Axes[0].grid(True)
```

```python
n= np.array([0,1,2,3,4,5])
fig,axes = plt.subplots(1,4,figsize=(16,5))
axes[0].set_title("scatter")
axes[0].scatter(x, x+0.25*np.random.randn(len(x)))
axes[1].set_title("step")
axes[1].step(n,n**2,lw=2)
axes[2].set_title("bar")
axes[2].bar(n,n**2, align="center")
axes[3].set_title("fill_between")
axes[3].fill_between(n,n**2,n**3,color="green", alpha=0.5)
```

```python
n=np.random.randn(100000)
fig,axes = plt.subplots(1,2,figsize=(12,5))
axes[0].set_title("default histogram")
axes[0].hist(n)
axes[1].set_title("Cumulative detailed histogram")
axes[1].hist(n, cumulative=True, bins=50)
```
